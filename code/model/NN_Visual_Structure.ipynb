```python
import matplotlib.pyplot as plt

def draw_neural_net(ax, layer_sizes, layer_colors=None):
    """
    Draw a simple neural network diagram, fully centered on both X and Y axes.

    Parameters:
    - ax: matplotlib axes
    - layer_sizes: list with number of neurons per layer
    - layer_colors: optional list of colors for each layer
    """
    n_layers = len(layer_sizes)
    h_spacing = 2  # horizontal spacing between layers

    # Horizontal centering offset
    total_width = (n_layers - 1) * h_spacing
    x_shift = -total_width / 2.0

    neuron_coords = []
    for i, layer_size in enumerate(layer_sizes):
        x = i * h_spacing + x_shift
        # Vertical centering for each layer individually
        total_height = (layer_size - 1)
        y_shift = total_height / 2.0
        layer_coords = []
        for j in range(layer_size):
            y = y_shift - j
            layer_coords.append((x, y))
            color = layer_colors[i] if layer_colors else 'skyblue'
            circle = plt.Circle((x, y), 0.2, color=color, ec='k', zorder=4)
            ax.add_patch(circle)
        neuron_coords.append(layer_coords)

    # Draw connections
    for (layer1, layer2) in zip(neuron_coords[:-1], neuron_coords[1:]):
        for (x1, y1) in layer1:
            for (x2, y2) in layer2:
                ax.plot([x1, x2], [y1, y2], 'k-', lw=0.5, zorder=1)

    # Turn off axes
    ax.axis('off')
    ax.set_aspect('equal')

# Example usage
fig, ax = plt.subplots(figsize=(10, 6))
layer_sizes = [3, 5, 4, 2]  # neurons per layer
layer_colors = ['lightgreen', 'skyblue', 'skyblue', 'lightcoral']
draw_neural_net(ax, layer_sizes, layer_colors)
plt.title("Neural Network Architecture")
plt.savefig(r"C:\Users\mvana\Documents\MSc\Temp_O3_H2O\Plots\NN_Architecture.png", format='png', dpi=1000, bbox_inches='tight')
plt.show()
```
